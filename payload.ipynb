```python
from fastapi import FastAPI
from pydantic import BaseModel
from typing import List, Optional

app = FastAPI()

class Penilaian(BaseModel):
    confidence_score: int
    kualitas_jawaban: int
    relevansi: int
    koherensi: int
    tempo_bicara: int
    total: int

class Result(BaseModel):
    penilaian: Penilaian
    penilaian_akhir: int
    cheating_detection: str
    alasan_cheating: Optional[str]
    analisis_non_verbal: str
    keputusan_akhir: str

class Interview(BaseModel):
    id: int
    result: Result

class Peserta(BaseModel):
    name: str
    content: List[Interview]
```

```python
@app.get("/evaluasi", response_model=Peserta)
def get_dummy_evaluasi():
    dummy_penilaian = Penilaian(
        confidence_score=94,
        kualitas_jawaban=100,
        relevansi=90,
        koherensi=80,
        tempo_bicara=100,
        total=90
    )
    dummy_result = Result(
        penilaian=dummy_penilaian,
        penilaian_akhir=5,
        cheating_detection="Tidak",
        alasan_cheating=None,
        analisis_non_verbal="Lancar dan tidak mencurigakan",
        keputusan_akhir="Lulus"
    )
    return Peserta(
        name="Naufal Ahmad",
        content=[
            Interview(id=1, result=dummy_result),
            Interview(id=2, result=dummy_result),
            Interview(id=3, result=dummy_result)
        ]
    )
```

```python
import nest_asyncio
from pyngrok import ngrok
import uvicorn

nest_asyncio.apply()
public_url = ngrok.connect(8888)
print("Public URL:", public_url)

config = uvicorn.Config(app=app, port=8888)
server = uvicorn.Server(config=config)
await server.serve()
```
